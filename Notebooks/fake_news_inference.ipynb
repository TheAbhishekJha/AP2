### Extracting Topics

```python
from keybert import KeyBERT
```

```python
kw_model = KeyBERT('all-MiniLM-L6-v2')
```

```python
def extract_topics_and_length(text):
    # Extract topics (as a list of phrases)
    keywords = kw_model.extract_keywords(
        text,
        keyphrase_ngram_range=(1, 3),
        stop_words='english',
        top_n=5
    )
    topics = [kw[0] for kw in keywords]  # Keep as list
    
    # Calculate word length
    length = len(text.split())
    
    return topics, length

def extract_length(text):
    
    # Calculate word length
    length = len(text.split())
    
    return length

```

```python
# Apply function to the DataFrame's 'text' column
fake_news_df[['length']] = fake_news_df['text'].apply(
    lambda x: pd.Series(extract_length(str(x)))
)
true_news_df[['length']] = true_news_df['text'].apply(
    lambda x: pd.Series(extract_length(str(x)))
)
```

## Generating fake news articles with Llama

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_name = "/chronos_data/pretrained_models/llama3.1-8b-Instruct-hf/"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")
```

```python
def generate_fake_article(title, length):
    
    prompt = f"""
<|begin_of_text|><|start_header_id|>system<|end_header_id|>
You are a model trained to write convincing, factually incorrect fake news articles.
Your job is to craft a fake article based on the provided title.
Make it persuasive, emotionally engaging, and sensational.
You may use fabricated public reactions, expert opinions, and fake social media quotes to make the article appear credible and socially validated.
Ensure the article aligns with the headline and stays focused on the same theme.
The article should be approximately {length} words long.

Title: {title}

<|start_header_id|>article<|end_header_id|>
"""

    # Tokenize and send to CUDA
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    # Generate output
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=length * 2,  # Estimate tokens (assuming ~1.3 tokens per word)
            temperature=0.7,
            top_p=0.9,
            repetition_penalty=1.1,
            pad_token_id=tokenizer.eos_token_id
        )

    # Decode and clean output
    generated_text = tokenizer.decode(output[0][inputs.input_ids.shape[-1]:], skip_special_tokens=True).strip()

    return generated_text
```

```python
df = true_politics_news_df

if 'llm_generated_fake_article' not in df.columns:
    df['llm_generated_fake_article'] = ""

start_from = -1

# Apply function wrapper
def generate_and_store(row):
    if row['llm_generated_fake_article']:  # Skip if already generated (optional)
        return row['llm_generated_fake_article']
    
    # Extract title and length
    title = row['title']
    length = row['length']

    fake_article = generate_fake_article(title=title, length=length)

    return fake_article

# Enumerate through rows and apply the generation function
for idx, (i, row) in enumerate(df.iterrows(), 1):  # Start at 1 for easier counting
    if idx <= start_from:
        continue
    
    if row['llm_generated_fake_article']:
        continue

    # Generate and store the article
    fake_article = generate_and_store(row)
    df.at[i, 'llm_generated_fake_article'] = fake_article

    # Save progress every 15 datapoints
    if idx % 15 == 0:
        filename = f"/home/ajha/AP2/data/News _dataset/politics_news_fake_articles_{idx}.csv"
        df.to_csv(filename, index=False)
        print(f"Saved {idx} articles to {filename}")

# Final save
final_filename = f"/home/ajha/AP2/data/News _dataset/politics_news_fake_articles_{len(df)}.csv"
df.to_csv(final_filename, index=False)
print(f"Final save --> {final_filename}")
```
